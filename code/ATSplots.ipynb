```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#load csv
df = pd.read_csv('/Users/zakg04/Documents/HBC Lab/BOOST/Main/test-7005/ATS/7005_ATS_A.csv')


#drop practice data
df = df[df['block'] == 'test']
```

```python
#store percent responded by block
# if block == none, then no response was made
percent_res = df['response'].apply(lambda x: 1 if x != 'None' else 0)
percent_res = percent_res.groupby(df['block_c']).mean()

percent_res
```

```python
#store percent correct by block
percent_corr = df['correct'].groupby(df['block_c']).mean()
percent_corr
```

```python
#store mean RT by block
mean_rt = df['response_time'].groupby(df['block_c']).mean()
mean_rt
```

```python
# plot percent responded by block as a barchart, scale should be 0-1
plt.figure()
percent_res.plot(kind='bar', title='Percent Responded by Block')
plt.ylabel('Percent Responded')
plt.xlabel('Block')
plt.ylim(0,1)
plt.show()
```

```python
# plot percent correct by block as a barchart, scale should be 0-1
plt.figure()
percent_corr.plot(kind='bar', title='Percent Correct by Block')
plt.ylabel('Percent Correct')
plt.xlabel('Block')
plt.ylim(0,1)
plt.show()
```

```python
#plot rt as a dot plot with block on the x-axis and rt on the y-axis
# add box plot with means and standard deviations
plt.figure()
sns.stripplot(x='block_c', y='response_time', data=df, jitter=True)
sns.boxplot(x='block_c', y='response_time', data=df, showcaps=False, boxprops={'facecolor':'None'}, showfliers=False)
plt.title('Reaction Time by Block')
plt.ylabel('Reaction Time')
plt.xlabel('Block')
plt.show()
```
